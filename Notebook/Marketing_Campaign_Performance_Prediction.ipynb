Now concatinating (merging) nykaa, purplle & tira dataframes into single dataframes

```python
# to combine the three dataframes, combined_df = pd.concat([df1,df2,df3], axis=0, ignore_index=True)

# axis = 0            => datas added vertically, records by records
# axis = 1            => datas added horizontally, features by features like adding new columns
# ignore_index = True => index from df1, df2, df3 will be ignored and new index will be added.

nykaa_purplle_tira_compaign_combined_dataframe = pd.concat([nykaa_compaign_dataframe,purplle_campaign_dataframe,tira_campaign_dataframe],axis=0, ignore_index=True)
print("Data shape:", nykaa_purplle_tira_compaign_combined_dataframe.shape)

# to find the count of null values in each columns
print("\nMissing values:")
print(nykaa_purplle_tira_compaign_combined_dataframe.isnull().sum())

# to remove the records which has null value in the column name is "Campaign_Type", "Channel_Used","Target_Audience"
# df.dropna(subset = ['column1', 'column2', 'column3'], inplace=True) 
# inplace = True => reflects in original dataframe
nykaa_purplle_tira_compaign_combined_dataframe = nykaa_purplle_tira_compaign_combined_dataframe.dropna(subset = ['Campaign_Type','Channel_Used','Target_Audience'])
print("\nData shape:", nykaa_purplle_tira_compaign_combined_dataframe.shape)

print("\nMissing values:")
print(nykaa_purplle_tira_compaign_combined_dataframe.isnull().sum())

print("\nStatistical Sumarries:")
print(nykaa_purplle_tira_compaign_combined_dataframe.describe())
```

Now split the Channel_Used columns into Channel_Used_Email, Channel_Used_Facebook,.......  

one-Hot Encoding

```python
# One Hot encoding:

# df.['column'].str.split(', ').str.join('|').str.get_dummies().add_prefix('word_')
# .str.split(', ')     =>  split the strings
# .str.join('|')       =>  join the strings
# .str.get_dummies()   =>  create individual dummy column and give values in numerical as '1' and '0'.
# .add_prefix('Channel_Used')  =>  function is used to add prefix in each column name. 
nykaa_purplle_tira_one_hot_encoded = nykaa_purplle_tira_compaign_combined_dataframe['Channel_Used'].str.split(', ').str.join('|').str.get_dummies().add_prefix('Channel_Used_')
print("One Hot Encoding process:")
print(nykaa_purplle_tira_one_hot_encoded.head())

# concatinate the original dataframe and encoded columns horizontally:
nykaa_purplle_tira_dataframe_encoded = pd.concat([nykaa_purplle_tira_compaign_combined_dataframe, nykaa_purplle_tira_one_hot_encoded], axis=1)
print("\nAfter Concatinating orginal dataframe and encoded columns:")

print("\nData Shape:", nykaa_purplle_tira_dataframe_encoded.shape)
print("\nBefore Rerranging the columns order:")
print(nykaa_purplle_tira_dataframe_encoded.head())

print("\nInfo to find index number for each columns:")
print(nykaa_purplle_tira_dataframe_encoded.info())

# find the column index number in the dataframe:
# index number = df.columns.get_loc('column_name')
channel_used_position = nykaa_purplle_tira_dataframe_encoded.columns.get_loc('Channel_Used')
print(f"\nIndex of Channel_Used: {channel_used_position}")


print("\nAfter Rearranging the columns order:")
colns = nykaa_purplle_tira_dataframe_encoded.columns.tolist()
rearrange_colns = colns[:6] + colns[17:] + colns[6:17]

nykaa_purplle_tira_dataframe_encoded_rearranged = nykaa_purplle_tira_dataframe_encoded[rearrange_colns]
print(nykaa_purplle_tira_dataframe_encoded_rearranged.head())


```

Data processing 

```python
# Data processing:

print(nykaa_purplle_tira_dataframe_encoded_rearranged.head())

# make a copy of orginal dataframe encoded rearranged:
nykaa_purplle_tira_dataframe_encoded_rearranged_copied = nykaa_purplle_tira_dataframe_encoded_rearranged.copy()
print("\nMissing values:")
print(nykaa_purplle_tira_dataframe_encoded_rearranged.isnull().sum())

print("\nInformation")
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.info())

print("\nMissing values:")
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.isnull().sum())

# Fill Missing value for Revenue, Acquistion cost:
# Fill zero for missing value in Revenue:
nykaa_purplle_tira_dataframe_encoded_rearranged_copied['Revenue'] = nykaa_purplle_tira_dataframe_encoded_rearranged_copied['Revenue'].fillna(0)
print("\nMissing values:")
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.isnull().sum())

# Fill mean value in Acquistion Cost:
mean_value = nykaa_purplle_tira_dataframe_encoded_rearranged_copied['Acquisition_Cost'].mean().round(2)
print(f"\ncalculated mean value: {mean_value}")
nykaa_purplle_tira_dataframe_encoded_rearranged_copied['Acquisition_Cost'] = (
    nykaa_purplle_tira_dataframe_encoded_rearranged_copied['Acquisition_Cost'].fillna(value=mean_value)
                                                                            )
print("\nMissing values:")
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.isnull().sum())

# Add new column 'Profit' and perform operation 'Revenue' - 'Acquistion_Cost' to assign a value:
# df = df.assign(New_Column_Name = lamda X: X['Column_1'] + X['Column_2'])
nykaa_purplle_tira_dataframe_encoded_rearranged_copied = (
    nykaa_purplle_tira_dataframe_encoded_rearranged_copied.assign(Profit=lambda X: X['Revenue'] - X['Acquisition_Cost'])
                                                            ) 

# Add new column 'New calculated ROI' and perform operation 'Profit' / 'Acquistion_Cost' to assign a value:
nykaa_purplle_tira_dataframe_encoded_rearranged_copied = (
                nykaa_purplle_tira_dataframe_encoded_rearranged_copied.assign(New_Calc_ROI=lambda X: X['Profit'] / X['Acquisition_Cost'])
                                                        )
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.info())

# Round off the decimal numbers upto two decimal point:
#df.round({"Column_1": 1, "Column_2": 2})
nykaa_purplle_tira_dataframe_encoded_rearranged_copied = (
     nykaa_purplle_tira_dataframe_encoded_rearranged_copied.round({"Profit": 2, "New_Calc_ROI": 2})
                                                        )

# Add new column 'New calculated ROI in percentage' and perform operation 'New_Calc_ROI' * 100 to assign a value:
nykaa_purplle_tira_dataframe_encoded_rearranged_copied = (
                nykaa_purplle_tira_dataframe_encoded_rearranged_copied.assign(New_Calc_ROI_Per=lambda X: X['New_Calc_ROI'] * 100)
                                                        )
print("\nMissing values:")
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.isnull().sum())
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.info())
print(nykaa_purplle_tira_dataframe_encoded_rearranged_copied.head())

nykaa_purplle_tira_dataframe_encoded_rearranged_copied.to_csv(r"C:\Users\jagad\Documents\my_classes\Tasks\mini_project_3-marketing_campaign_performance_prediction\rough_data\nykaa_purplle_tira_combined_files\nykaa_purplle_tira_dataframe_encoded_rearranged_copied_2.csv")
```
